```python
import pandas_datareader.data as web
from datetime import datetime, timedelta
import pandas as pd
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import numpy as np
import yfinance as yf
```

```python
tickerSymbol = "LLOY.L"
```

```python
tickerData = yf.Ticker(tickerSymbol)
```

```python
tickerDf = tickerData.history(interval = '1d', 
                             start = '2015-01-01',
                             end  = '2021-01-10')
```

```python
priceData = tickerDf.Close
```

```python
priceData = priceData.asfreq(pd.infer_freq(priceData.index))
```

```python
priceData
```

```python
plt.figure(figsize = (10,4))
plt.plot(priceData, color = 'red')
for year in range(priceData.index[0].year, priceData.index[-1].year+1):
    plt.axvline(datetime(year,1,1), color='blue', linestyle='--', alpha=0.75)
plt.title("%s Price Data"%tickerSymbol, fontsize=20, color = 'white')
plt.ylabel("",color = 'white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
```

```python
returns = priceData.pct_change().dropna()
```

```python
plt.figure(figsize=(10,4))
plt.plot(returns, color = 'black')
plt.ylabel('Pct Return', fontsize=16)
plt.title('%s Squared Returns'%tickerSymbol, fontsize=20, color = 'white')
plt.ylabel("",color = 'white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
```

# PACF

```python
plot_pacf(returns**2, color = 'green', method = 'ywm', lags = 20)
plt.title("PACF", color = 'white', size = 20)
plt.ylabel("",color = 'white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')
plt.show()
```

```python
model = arch_model(returns, p=1, q=1,dist = 'ged', mean = 'AR', rescale = False).fit()
```

```python
from arch.__future__ import reindexing
rolling_predictions = []
test_size = 365

for i in range(test_size):
    train = returns[:-(test_size-i)]
    model = arch_model(train**2, p=1, q=1, rescale=False, dist = 'studentst', mean = 'AR')
    model_fit = model.fit(disp='true')
    pred = model_fit.forecast(horizon=1)
    rolling_predictions.append(np.sqrt(pred.variance.values[-1,:][0]))
```

```python
rolling_predictions = pd.Series(rolling_predictions, index=returns.index[-365:])
```

```python
plt.figure(figsize=(14,4))
true, = plt.plot(returns[-365:], color = 'red')
preds, = plt.plot(rolling_predictions, color = 'black')
invPred, = plt.plot(-1*rolling_predictions, color = 'black')
plt.title('Volatility Prediction - Rolling Forecast', fontsize=20)
plt.legend(['True Returns', 'Predicted Volatility'], fontsize=16)
```
